# Phos vs Temp

## Maxium

```python
Max_data= pd.read_csv('Max_heatmap_data.csv')
```

```python
Max_data
```

```python
Max_data=Max_data.sort_values(by='Unnamed: 0', ascending=False)
Max_data
```

```python
Temp = np.array(np.array(Max_data['Unnamed: 0']))
Pin = [float(Max_data.columns[i]) for i in range(1,Max_data.shape[1],1)]
Max_matrix = Max_data[Max_data.columns[1:]]
```

```python
map_color = 'inferno'
ax = sns.heatmap(Max_matrix, cmap=map_color)
ax.set(xlabel="$p_{in}$ $[mgP/L]$",
           ylabel="Increase Temp $\degree C$")

x_ticks = ax.xaxis.get_ticklocs()
if len(Pin) > len(x_ticks):
        x_stape = round(len(Pin) / len(x_ticks), 0)
        x_labels = [Pin[int(a*x_stape)] for a in range(len(x_ticks))]
        ax.xaxis.set_ticklabels(x_labels,
                                rotation=90)
else:
        x_labels = Pin
        ax.xaxis.set_ticklabels(x_labels,
                                rotation=90)

y_ticks = ax.yaxis.get_ticklocs()
    # Z_k_list = list(self.Z_k)
    # step = round(len(self.Z_k)/len(y_ticks))
    # y_labels = [Z_k_list[a*step] for a in range(len(y_ticks))]
if len(Temp) > len(y_ticks):
        y_stape = round(len(Temp) / len(y_ticks), 0)
        y_labels = [Temp[int(a*y_stape)] for a in range(len(y_ticks))]
        ax.yaxis.set_ticklabels(y_labels,
                                rotation=0)
else:
        y_labels = Temp
        ax.yaxis.set_ticklabels(y_labels,
                                rotation=0)
# Adding label to the colorbar
cbar = ax.collections[0].colorbar
cbar.set_label('Microcystin-LR ($\mu$g/L)', rotation=90)
# Save plots
# path = './New data/Images/Year v_1/'
# name = 'Full_model_heat_MaxM_pho_max.png'
# plt.savefig(path+name, dpi=500, bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.ticker import MaxNLocator
from matplotlib.ticker import MaxNLocator
import matplotlib.pyplot as plt
from matplotlib.colors import BoundaryNorm
from matplotlib.colorbar import ColorbarBase
import pandas as pd
import numpy as np
```

```python
def heatmap_max_value(custom_cmap = 'turbo', figsize=(12, 6), FONTSIZE=7):
    path_temp = ['0.6', '1.0', '1.4', '0.4', '1.7','3.4']
    max_lim = 0
    min_lim = 100
    for TEMP in path_temp:
        Mean_data= pd.read_csv('SSP2_dE_zm_max_maximumValue_'+TEMP+'.csv')
        Mean_data = Mean_data[Mean_data.columns[1:]]
        max_lim = max(max_lim, Mean_data.max().max())
        min_lim = min(min_lim, Mean_data.min().min())

    custom_cmap = custom_cmap
    # Define custom colormap colors
    num_levels_color = 7 * 3
    # Define number classes
    num_levels_line = 7 * 1
    
    
    fig, axs = plt.subplots(2, 3, figsize= figsize,layout='constrained')
    axs = axs.ravel()
    
    for i, TEMP in enumerate(path_temp):
        Mean_data = pd.read_csv('SSP2_dE_zm_max_maximumValue_' + TEMP + '.csv')
        Mean_matrix = Mean_data.iloc[:, 1:].values
        Z_values = np.arange(1, 10+0.1, 0.05).round(2)
        d_values = np.arange(0.02, 0.08 + 0.0005, 0.0005).round(4)
        MeanFrame = pd.DataFrame(Mean_matrix, index=d_values, columns=Z_values)
    
        x = Z_values
        y = d_values[::-1]
    
        X, Y = np.meshgrid(x, y)
    
        Z = Mean_matrix
    
        # Plot heatmap with contour lines using the custom colormap
        contour_levels_colors = np.linspace(Z.min().min(), Z.max().max(), num_levels_color + 1)
        contour_levels_colors = np.linspace(min_lim, max_lim, num_levels_color + 1)
        pcm = axs[i].contourf(X, Y, Z, levels=contour_levels_colors, cmap=custom_cmap, vmin=min_lim, vmax=max_lim)
        # axs[i].contour(X, Y, Z, levels=num_levels_line, colors='k', alpha=0.7)
    
        axs[i].set_xlabel("Epilimnion depth", fontsize=FONTSIZE)  # Use set_xlabel instead of xlabel
        axs[i].set_ylabel("Water exchange rate", fontsize=FONTSIZE)  # Use set_ylabel instead of ylabel
    
        # Set ticks and labels for x and y axes
        print_Z = list(np.around(np.arange(1, 10 , 0.7*2), 2))
        # print_Z.append(10)
        print_d = list(np.around(np.arange(0.02, 0.08 + 0.005, 0.005*2), 3))
        print_d.append(0.08)
        print_d.reverse()
        axs[i].xaxis.set_ticks(print_Z)  # Use set_ticks instead of set_ticklabels
        axs[i].xaxis.set_ticklabels(print_Z, rotation=90, fontsize=FONTSIZE)
        axs[i].yaxis.set_ticks(print_d)  # Use set_ticks instead of set_ticklabels
        axs[i].yaxis.set_ticklabels(print_d, fontsize=FONTSIZE)
    
    axs[0].set_title('SSP1-1.9', fontsize=FONTSIZE+2, pad = 10, fontweight='bold')
    axs[1].set_title('SSP2-4.5', fontsize=FONTSIZE+2, pad = 10, fontweight='bold')
    axs[2].set_title('SSP5-8.5', fontsize=FONTSIZE+2, pad = 10, fontweight='bold')

    fig.text(-0.025,  0.770, 'Medium-Term', va='center',
             rotation='vertical', fontsize=FONTSIZE+2, 
             fontweight='bold', alpha=1)
    fig.text(-0.025, 1-0.725, 'Long-Term', va='center', rotation='vertical', 
             fontsize=FONTSIZE+2, fontweight='bold', alpha=1)
    # fig.suptitle('Vulnerability index with respect to maximum microcystin-LR concentration', 
    #                 fontsize=FONTSIZE+2,fontweight='bold')
    # Create a colorbar
    cbar_ax = fig.add_axes([1.005, 0.085, 0.025, 0.875])  # [left, bottom, width, height]
    cbar = fig.colorbar(pcm, cax=cbar_ax)
    cbar.ax.tick_params(labelsize=FONTSIZE) 
    cbar.set_label('Vulnerability index', fontsize=FONTSIZE)
    
    # plt.tight_layout()
    
    return fig, axs
```
