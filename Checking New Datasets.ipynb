# Creating Data, number of samples for lakes from 2018 to 2023

```python
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv("merged_water_quality_data_V2.csv",low_memory=False)
```

```python
Stations = [station for station in dataset['Station'].unique() if station.endswith('COMPOSITE') or 'PROFILE' in station
                           or station.endswith('(OLD)')]
```

```python
year = '2017'
dataset['SampleDate'] = pd.to_datetime(
        dataset['SampleDate'], format='%Y/%m/%d')
year_start = pd.Timestamp(year=int(year), month=1, day=1)
year_end = pd.Timestamp(year=int(year) + 1, month=1, day=1)
C1 = dataset['Station'].str.contains('pine lake'.upper())
Con2 = dataset['SampleDate'] >= year_start
Con3 = dataset['SampleDate'] < year_end
dataset[C1 & Con2&Con3][['Total cyanobacterial cell count (cells/mL)','SampleDate' ]].dropna()
```

```python
len(Stations)
```

```python
dataset['SampleDate'] = pd.to_datetime(dataset['SampleDate'] , format='%Y/%m/%d')
```

```python
C1 = dataset['SampleDate']>= pd.to_datetime('2016')
C2 = dataset['SampleDate']< pd.to_datetime('2017')
```

```python
samplesStation = {}
for station in Stations:
    C3 = dataset['Station'] == station
    samplesStation[station] = dataset[C1 & C2 & C3].count()
    
```

```python
samplesStation =pd.DataFrame(samplesStation)
```

# US  Lakes

```python
import pandas as pd
```

```python
dataUs = pd.read_csv("Combined_Data_for_MO.csv")
```

```python
dataUs['sampledate'] = pd.to_datetime(dataUs['sampledate'])
```

```python
dataUs.columns
```

```python
labels = ['sampledate', 'cells_per_ml', 'wtemp', 'o2']
grouped_data = dataUs[labels].groupby('sampledate').mean()
```

```python
grouped_data.rename(columns={'cells_per_ml': 'Total cyanobacterial cell count (cells/mL)', 
                   'o2': 'OXYGEN DISSOLVED (FIELD METER)',
                  'wtemp':'TEMPERATURE WATER'}, inplace=True)
grouped_data
```

```python
grouped_data.to_csv("Combined_Data_for_MO_Merged.csv")
```
